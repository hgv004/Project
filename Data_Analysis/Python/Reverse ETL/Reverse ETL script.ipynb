```python
import pandas as pd
import os
```

```python
path = r"C:\Hardik\Projects\Data_Analysis\Python\Reverse ETL\Datasource\all data04-07.xlsx"
```

```python
df = pd.ExcelFile(path)
df.sheet_names
```

```python
data = pd.DataFrame()
for i in df.sheet_names :
    df2 = pd.read_excel(path, sheet_name = i)
    print(df2)
    data = pd.concat([data, df2], ignore_index=True)
print(data)
```

```python
op_path = r"C:\Hardik\Projects\Data_Analysis\Python\Reverse ETL\Output"
```

```python
for date in data['Date'].unique():
    # Filter the DataFrame for the current date
    filtered_df = data[data['Date'] == date]
    
    # Define the output file path based on the date
    output_file = f"{date.strftime('%Y-%m-%d')}.csv"
    
    # Save the filtered DataFrame to a CSV file
    filtered_df.to_csv(op_path + "\\" + output_file, index=False)
    
    print(f"Saved {output_file}")
```

```python
# Create a directory to store the files
output_dir = op_path
os.makedirs(output_dir, exist_ok=True)

# Iterate through unique dates
for date in data['Date'].unique():
    # Filter the DataFrame for the current date
    filtered_df = data[data['Date'] == date]
    
    # Define the year, month, and day
    year = date.strftime('%Y')
    month = date.strftime('%m')
    day = date.strftime('%d')
    
    # Define the output directory path based on year and month
    output_subdir = os.path.join(output_dir, year, month)
    os.makedirs(output_subdir, exist_ok=True)
    
    # Define the output file path based on the date
    output_file = os.path.join(output_subdir, f"{year}-{month}-{day}.csv")
    
    # Save the filtered DataFrame to a CSV file
    filtered_df.to_csv(output_file, index=False)
    
    print(f"Saved {output_file}")
```
