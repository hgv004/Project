# MYSQL Connection

```python
db_url = r'mysql+pymysql://root:root@localhost:3306/etl'
```

```python
engine = create_engine(db_url)
mydb = engine.connect()
```

```python
f1 = pd.read_csv(r"C:\Hardik\Projects\Data_Analysis\Python\ETL Pipeline\Datasource\f1\2023-09-15.csv")
f2 = pd.read_csv(r"C:\Hardik\Projects\Data_Analysis\Python\ETL Pipeline\Datasource\f2\2023-09-15.csv")
f3 = pd.read_csv(r"C:\Hardik\Projects\Data_Analysis\Python\ETL Pipeline\Datasource\f3\2023-09-15.csv")
```

```python
f1['updated_date'] = '2023-09-15'
f2['updated_date'] = '2023-09-15'
f3['updated_date'] = '2023-09-15'
```

```python
f1.to_sql(name='f1', con=engine, if_exists="replace", index=False)
f2.to_sql(name='f2', con=engine, if_exists="replace", index=False)
f3.to_sql(name='f3', con=engine, if_exists="replace", index=False)
```

```python
# Define your SQL query
q = text("select max(updated_date) from f1")

# Execute the query and fetch the results
with engine.connect() as connection:
    result = connection.execute(q)
    latest_date_db = result.fetchall()[0][0]
    print(latest_date_db)
```

```python
datadict = {}

# Prepare function to combine sftp files dataframe
def comb_data(folder, db_table):
    # Access local files
    path = os.listdir(fpath + f"/" + folder)
    Local_filelist = pd.Series(path).str[:-4]
    print(Local_filelist)

    # Extract file with latest date
    maxdate_local = datetime.strptime(max(Local_filelist), "%Y-%m-%d")
    maxdate_local = maxdate_local.date()
    print(maxdate_local)
    
    # Define your SQL query
    q = text(f"SELECT max(updated_date) FROM {db_table}")
    print(q)

    # Execute the query and fetch the results
    with engine.connect() as connection:
        result = connection.execute(q)
        latest_date_db = result.fetchall()[0][0]
        print(latest_date_db)

    # Create DataFrame for new files
    Local_filelist_date = pd.to_datetime(Local_filelist, format="%Y-%m-%d")
    Local_filelist_date = Local_filelist_date.dt.date

    files_to_append = []
    for i in Local_filelist_date:
        if i > latest_date_db:
            files_to_append.append(i.strftime("%Y-%m-%d") + ".csv")
            print(type(i), type(latest_date_db))
    print(files_to_append)

    data = pd.DataFrame()
    for i in files_to_append:
        local_filepath = fpath + f"\\" + 'f1'
        df1 = pd.read_csv(local_filepath + f"\\" + i, index_col=False)
        df1["updated_date"] = i[:-4]
        data = pd.concat([data, df1], ignore_index=True)
    return data
    
tables = ['f1', 'f2', 'f3']

for i in tables:
    datadict[i] = comb_data(i, i)

for i, j in datadict.items():
    j.to_sql(name=i, con=engine, if_exists="append", index=False)
```
